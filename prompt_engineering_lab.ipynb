# Part 1: Starting Simple - Create Initial Prompts

## Objective: Create basic zero-shot prompts for three different task types and establish a baseline.

### Step 1: Set Up the Environment

```python
# Import required libraries
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python

# Load environment variables from .env file
load_dotenv()

# Initialize OpenAI client
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

# Set default model
MODEL = "gpt-4o-mini"

print("✅ Setup complete! OpenAI client initialized.")
```

```python
def call_openai(prompt, system_message="You are a helpful assistant.", temperature=0.7, max_tokens=None, seed=None):
    """Helper function to call OpenAI API with specified parameters"""
    try:
        params = {
            "model": MODEL,
            "messages": [
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt}
            ],
            "temperature": temperature
        }
        
        if max_tokens:
            params["max_tokens"] = max_tokens
        if seed is not None:
            params["seed"] = seed
            
        response = client.chat.completions.create(**params)
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

def display_response(title, response, params=None):
    """Display a response with formatting"""
    print(f"\n{'='*60}")
    print(f"📝 {title}")
    if params:
        print(f"Parameters: {params}")
    print(f"{'='*60}")
    print(response)
    print(f"{'='*60}\n")

def count_tokens_approx(text):
    """Approximate token count (roughly 4 characters per token)"""
    return len(text) // 4

print("✅ Helper functions loaded!")
```

# Part 2: Diagnosing Failures - Systematic Testing

## Objective: Run each prompt multiple times (5, 10, 15 iterations) to identify failure patterns and inconsistencies.

### Step 3: Run Prompts 5 Times

### Project Tasks

- Run each prompt 5 times using a helper function
- Record all responses
- Analyze consistency between prompts

```python
prompts = {
    "Sentiment Analysis": sentiment_prompt_v1,
    "Product Description": product_prompt_v1,
    "Data Extraction": extraction_prompt_v1
}

def run_single_prompt_n_times(prompt, n=5):
    """Run one prompt multiple times and return the responses."""
    responses = []

    for i in range(n):
        response = call_openai(prompt)
        responses.append(response.strip())

        print(f"\nRun {i+1}")
        print(f"{'-'*20}")
        print(response.strip())

    return responses

def run_prompt_n_times(prompts, n=5):
    """Run each prompt multiple times, print the responses, and return all results."""
    all_results = {}

    for name, prompt in prompts.items():
        print(f"\n{'='*60}")
        print(name)
        print(f"{'-'*60}")
        print(prompt.strip())
        print(f"{'-'*60}")

        all_results[name] = run_single_prompt_n_times(prompt, n)

    return all_results

results_5 = run_prompt_n_times(prompts, 5)
```

### Observations:

After running each prompt 5 times, it became clear that the outputs were not equally reliable across tasks. While all three prompts produced relevant answers, the level of consistency varied depending on the type of task.

**Are responses consistent in format?**  
Not fully. The responses followed a similar general structure, but the formatting changed between runs. Some outputs used bold text, different labels, different capitalization, or slightly different layouts.

**Are responses consistent in content?**  
Only partially. The overall meaning remained similar, but the exact wording changed from one run to another. In the sentiment task, the message was always interpreted positively, but the model used different labels. In the product description task, the model varied the product name, features, and level of detail. In the data extraction task, the main facts remained the same, but the labels and presentation changed.

**What types of variations do you see?**  
The main variations appeared in wording, formatting, response length, field naming, and level of detail. In the product description task, the model also introduced invented product names and additional features that were not provided in the original prompt.

**Consistency across prompts**  
Among the three tasks, the data extraction prompt showed the highest consistency because it repeatedly captured the same core facts, even with some formatting differences. The sentiment analysis prompt was moderately consistent, since the message was always recognized as positive, but the classification labels were not stable. The product description prompt showed the greatest variability, with noticeable differences in structure, tone, length, and generated product details.

Version 1 showed variation in field names, value formatting, and overall structure, even when the extracted facts were mostly correct. ### Interpretation

This step shows that Chain-of-Thought prompting can support structured analytical tasks by guiding the model through intermediate reasoning steps before producing the final answer. In this example, the task was simple enough that Version 2 had already solved the consistency problem, so Version 3 mainly reinforces the extraction logic rather than improving the final stability further.
