# Part 1: Starting Simple - Create Initial Prompts

## Objective: Create basic zero-shot prompts for three different task types and establish a baseline.

### Step 1: Set Up the Environment

```python
# Import required libraries
import os
from openai import OpenAI
from dotenv import load_dotenv
import json
import re
from collections import Counter
```

```python

# Load environment variables from .env file
load_dotenv()

# Initialize OpenAI client
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

# Set default model
MODEL = "gpt-4o-mini"

print("✅ Setup complete! OpenAI client initialized.")
```

```python
def call_openai(prompt, system_message="You are a helpful assistant.", temperature=0.7, max_tokens=None, seed=None):
    """Helper function to call OpenAI API with specified parameters"""
    try:
        params = {
            "model": MODEL,
            "messages": [
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt}
            ],
            "temperature": temperature
        }
        
        if max_tokens:
            params["max_tokens"] = max_tokens
        if seed is not None:
            params["seed"] = seed
            
        response = client.chat.completions.create(**params)
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

def display_response(title, response, params=None):
    """Display a response with formatting"""
    print(f"\n{'='*60}")
    print(f"📝 {title}")
    if params:
        print(f"Parameters: {params}")
    print(f"{'='*60}")
    print(response)
    print(f"{'='*60}\n")

def count_tokens_approx(text):
    """Approximate token count (roughly 4 characters per token)"""
    return len(text) // 4

print("✅ Helper functions loaded!")
```

# Part 2: Diagnosing Failures - Systematic Testing

## Objective: Run each prompt multiple times (5, 10, 15 iterations) to identify failure patterns and inconsistencies.

### Step 3: Run Prompts 5 Times

### Project Tasks

- Run each prompt 5 times using a helper function
- Record all responses
- Analyze consistency between prompts

```python
prompts = {
    "Sentiment Analysis": sentiment_prompt_v1,
    "Product Description": product_prompt_v1,
    "Data Extraction": extraction_prompt_v1
}

def run_single_prompt_n_times(prompt, n=5):
    """Run one prompt multiple times and return the responses."""
    responses = []

    for i in range(n):
        response = call_openai(prompt)
        responses.append(response.strip())

        print(f"\nRun {i+1}")
        print(f"{'-'*20}")
        print(response.strip())

    return responses

def run_prompt_n_times(prompts, n=5):
    """Run each prompt multiple times, print the responses, and return all results."""
    all_results = {}

    for name, prompt in prompts.items():
        print(f"\n{'='*60}")
        print(name)
        print(f"{'-'*60}")
        print(prompt.strip())
        print(f"{'-'*60}")

        all_results[name] = run_single_prompt_n_times(prompt, n)

    return all_results

results_5 = run_prompt_n_times(prompts, 5)
```

### Observations:

After running each prompt 5 times, it became clear that the outputs were not equally reliable across tasks in this setup. All three prompts produced relevant answers, but the level of exact-match consistency varied depending on the type of task and the amount of structure in the prompt.

**Are responses consistent in format?**  
Not fully. The responses followed a similar general structure, but the formatting changed between runs. Some outputs used bold text, different labels, different capitalization, or slightly different layouts.

**Are responses consistent in content?**  
Only partially. The overall meaning remained similar, but the exact wording changed from one run to another. In the sentiment task, the message was always interpreted positively, but the model used different labels. In this setup, that result should be interpreted carefully because the sentiment example is clearly positive and simple enough that deeper prompting strategies are not yet required. In the product description task, the model varied the product name, features, and level of detail. In the data extraction task, the main facts remained the same, but the labels and presentation changed.

**What types of variations do you see?**  
The main variations appeared in wording, formatting, response length, field naming, and level of detail. In the product description task, the model also introduced invented product names and additional features that were not provided in the original prompt.

**Consistency across prompts**  
Among the three tasks, the data extraction prompt showed the highest consistency because it repeatedly captured the same core facts, even with some formatting differences. The sentiment analysis prompt was moderately consistent, since the message was always recognized as positive, but the classification labels were not stable. However, because the sentiment example is very clearly positive, more experimentation with harder and more ambiguous cases will be needed in Step 12 to evaluate how much the later prompting methods really help. The product description prompt showed the greatest variability, with noticeable differences in structure, tone, length, and generated product details.


```python
# Automated comparison helpers and 5-run summary
def normalize_response(text):
    """Normalize whitespace and casing for lightweight comparisons."""
    return re.sub(r"\s+", " ", text.strip()).lower()

def safe_parse_json(text):
    """Parse JSON responses when possible, otherwise return None."""
    try:
        return json.loads(text)
    except Exception:
        return None

def summarize_runs(responses):
    """Return simple consistency metrics for a list of model responses."""
    normalized = [normalize_response(response) for response in responses]
    counts = Counter(normalized)
    most_common_output, most_common_count = counts.most_common(1)[0]

    return {
        "total_runs": len(responses),
        "unique_outputs": len(counts),
        "exact_match_rate": most_common_count / len(responses),
        "most_common_output": most_common_output,
    }

def print_run_summary(title, responses):
    """Print a compact summary for repeated runs of the same prompt."""
    summary = summarize_runs(responses)
    print(f"\n{title}")
    print("-" * len(title))
    print(f"Total runs: {summary['total_runs']}")
    print(f"Unique outputs: {summary['unique_outputs']}")
    print(f"Exact-match consistency: {summary['exact_match_rate']:.0%}")

def compare_prompt_sets(label_a, results_a, label_b, results_b):
    """Compare two dicts of prompt-run results."""
    print(f"\nComparison: {label_a} vs {label_b}")
    print("=" * 60)

    for task in results_a:
        summary_a = summarize_runs(results_a[task])
        summary_b = summarize_runs(results_b[task])
        delta = summary_b['exact_match_rate'] - summary_a['exact_match_rate']

        print(f"\n{task}")
        print(f"- {label_a}: {summary_a['unique_outputs']} unique outputs, {summary_a['exact_match_rate']:.0%} exact-match consistency")
        print(f"- {label_b}: {summary_b['unique_outputs']} unique outputs, {summary_b['exact_match_rate']:.0%} exact-match consistency")
        print(f"- Change in consistency: {delta:+.0%}")

for task_name, task_results in results_5.items():
    print_run_summary(f"5-run summary: {task_name}", task_results)
```

### Final Interpretation

The final comparison shows that the strongest improvements in this notebook came from adding explicit structure and tighter output constraints, not from the later advanced prompting techniques by themselves. In this setup, **sentiment analysis** improved from low format consistency in Version 1 to full consistency in Versions 2 and 3, but that result must be interpreted carefully because the original sentiment example was simple and clearly positive. The Step 12 variation tests gave a more realistic evaluation, showing that once the reviews became more ambiguous, the sentiment prompt was less robust and sometimes even broke the intended label set.

For **data extraction**, the most important change was the move to a fixed JSON schema. That change produced a large consistency gain in this setup, and Chain-of-Thought did not add measurable improvement on top of it for the base example. For **product description**, the prompts became more controlled in format, but exact-match consistency remained low because the task still allowed substantial wording variation and the model continued to generate different persuasive phrasing across runs.

Overall, this lab shows that prompt engineering can greatly improve reliability when the task is simple enough and the output format is tightly constrained. At the same time, the results also show why broader experimentation matters: when the original input is too easy, it becomes difficult to evaluate how much methods like few-shot prompting or Chain-of-Thought are really contributing. That is why the variation section is important for interpreting the sentiment results more realistically.
